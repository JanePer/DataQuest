## What's the most common median salary range?

```python
recent_grads['Median'].hist(bins=16, range=(0,80000))
```

```python
recent_grads["Median"].plot.box(title="Distribution of median salaries", 
                                grid=True)
```

The most common median year-round salary range is between 30K-45K, that is 64% of all majors.

## Do students in more popular majors make more money?

```python
recent_grads['Sample_size'].hist(bins=30, range=(0,3000))
```

```python
recent_grads["Sample_size"].plot.box(title="Distribution of Sample_size", 
                                grid=True)
```

78% of majors are represented with samples up to 400 respondents. As more os total number of graduates the bigger is sample size. This way the most popular majors have 500+ sample size.

```python
recent_grads.plot(x='Sample_size', y='Median', kind='scatter', xlim=(-100, 5000))
```

```python
from pandas.plotting import scatter_matrix
scatter_matrix(recent_grads[['Sample_size', 'Median']], alpha=0.8, figsize=(6,6))
```

Based on 2 above plots we can say that the more popular majors do not always mean more money, rather even vice versa, the most paid majors above than 70K are unpopular with the most less number of students. The most popular majors are set within the average year-round salary between 30K-50K.

## Is there any link between the number of full-time employees and median salary?

```python
recent_grads.plot(x='Full_time', y='Median', kind='scatter', xlim=(-10000, 260000), ylim=(15000, 120000))
```

```python
recent_grads.plot.hexbin(x='Full_time', y='Median', ylim=(15000, 120000), 
                         xlim=(-10000, 260000), gridsize=25)
```

Generally we can notice some trend where the bigger number of full-time employees is, the lower is median salary, while the majors with low number of full-time employees have such as the lowest salary as well as the highest. But the highest salary about more than 60K demonstrate majors it is always the majors with lower number of full-time employees.

## What percent of majors are predominantly male? Predominantly female?

```python
recent_grads['ShareWomen'].hist(bins=10, range=(0,1))
```

Let's assume that predominantly male majors are those which have ShareWomen less than 40% so predominantly female will be with ShareWomen more than 60%. Based on this logic there are: 
- 32% of predominantly male majors and 
- 42% of predominantly female majors.

### TOP 20 Female Majors

```python
fem_majors=recent_grads[recent_grads["ShareWomen"]>0.6].sort_values("ShareWomen", ascending=False)
fem_majors[:20].plot.bar(x='Major', y='ShareWomen', figsize=(15,5), grid=True)
```

### TOP 20 Male Majors

```python
male_majors=recent_grads[recent_grads["ShareWomen"]<0.4].sort_values("ShareWomen", ascending=True)
male_majors[:20].plot.bar(x='Major', y='ShareWomen', figsize=(15,5), grid=True)
```

## Grouped bar plot to compare the number of men with the number of women in each category of majors

```python
categories=recent_grads["Major_category"].value_counts().index
categories
```

```python
men_women_categories=[]
for c in categories:
    selected = recent_grads.loc[recent_grads["Major_category"]==c, ["Men", "Women"]]
    m=selected["Men"].sum()
    w=selected["Women"].sum()
    men_women_categories.append([c, m, w])
men_women_categories
```

```python
import pandas as pd
men_women_categories = pd.DataFrame(men_women_categories)
new_columns=["Major_category", "Sum_Men", "Sum_Women"]
men_women_categories.columns=new_columns
men_women_categories
```

```python
men_women_categories.plot.bar(x="Major_category", figsize=(10,5), 
              title="Number of men and women in each category of majors",
             grid=True)
```
